```python
latest_data=df[df['date']>"2021-02-28"] #viewing data for march-2021
```

```python
latest_data=latest_data.sort_values("date")#sorting the values by date
```

```python
plt.figure(figsize=(21,10))
plt.title("Report of March month")
sns.barplot(x="date",y="death",data=latest_data)#ploting   by seaborn
plt.figure(figsize=(21,10))
x_axis=np.arange(len(latest_data["date"]))

date=latest_data["date"]
death=latest_data["death"]
plt.title("Covid-19 Report Datewise")
plt.ylabel("Death")
plt.xlabel("Year ")

plt.bar(x_axis -0.2,death,width=0.3,label='date',color='red')# ploting by matplotlib
plt.xticks(x_axis,date)


plt.show()
```
